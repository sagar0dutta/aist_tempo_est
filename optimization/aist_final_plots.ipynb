```python

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), sharey=False)

# Titles and accuracy lists for each subplot
titles = ["Position-Based Accuracy", "Velocity-Based Accuracy", "Combined Position+Velocity Accuracy"]
acc_lists = [[], [], []]  # [Position, Velocity, Pos+Vel]


for w_sec in win_size_list:
    hop_size_list = [w_sec * 0.25, w_sec * 0.5, w_sec * 0.75, w_sec * 1]
    
    pos_hopacc_list = []
    vel_hopacc_list = []
    posvel_hopacc_list = []
    
    for h_sec in hop_size_list:
        hpos_hits_stats = load_from_pickle(f"./saved_results/{case}/stats_pos/tempo_{a}_{b}/{mode}/hit_stats_W{w_sec}H{h_sec}_{mode}_{a}_{b}.pkl")
        hvel_hits_stats = load_from_pickle(f"./saved_results/{case}/stats_vel/tempo_{a}_{b}/{mode}/hit_stats_W{w_sec}H{h_sec}_{mode}_{a}_{b}.pkl")
        hposvel_hits_stats = load_from_pickle(f"./saved_results/{case}/stats_posvel/tempo_{a}_{b}/{mode}/hit_stats_W{w_sec}H{h_sec}_{mode}_{a}_{b}.pkl")

        pos_hopacc_list.append(hpos_hits_stats[bpm_metric]['hits_summary']['hits_percentage'][4])  
        vel_hopacc_list.append(hvel_hits_stats[bpm_metric]['hits_summary']['hits_percentage'][4])
        posvel_hopacc_list.append(hposvel_hits_stats[bpm_metric]['hits_summary']['hits_percentage'][4])
    
    # Store results for each method
    acc_lists[0].append((w_sec, pos_hopacc_list))       # Position-Based
    acc_lists[1].append((w_sec, vel_hopacc_list))       # Velocity-Based
    acc_lists[2].append((w_sec, posvel_hopacc_list))    # Combined Pos+Vel

# colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]
cmap = plt.colormaps["tab20"] 
colors = [cmap(i / 13) for i in range(13)]
# Plot each method in separate subplots
for i, ax in enumerate(axes):
    for j, (w_sec, acc_values) in enumerate(acc_lists[i]):
        color = colors[j % len(colors)]
        ax.plot([0.25, 0.5, 0.75, 1], acc_values, marker='o', linestyle='-', label=f"Win:{w_sec} sec", color= color)
    
    ax.set_title(titles[i], fontsize=12)
    ax.set_xlabel("Normalized Hop Size (Window Overlap) ")
    ax.set_xticks(np.arange(0,1.25,0.25))
    ax.set_ylim(50, 100)  # Keep y-axis consistent
    ax.grid(True, which= 'both', linestyle='--', linewidth=0.5, alpha=0.7)
    ax.minorticks_on()
    
    if i == 0:
        ax.set_ylabel("Accuracy (%)")  # Add y-label only for first subplot
        ax.legend(fontsize=8, framealpha=0.5)  # Add legend


plt.suptitle("Accuracy vs. Hop Size Across Different Window Sizes", fontsize=14)
plt.tight_layout()
plt.show()
```
